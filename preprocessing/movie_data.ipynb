```python
import pandas as pd
```

```python
movies_df = pd.read_csv('../datasets/movies.csv')
```

```python
movies_df
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)   # create a separate 'year' column and extract year from title

movies_df
```

```python
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '', regex=True)  
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())  # remove years and possible whitespace from 'title' column

movies_df
```

```python
movies_df = movies_df.drop(columns='genres')
movies_df
```

```python
movies_df.to_csv("../datasets/movie_data.csv", index=False)
```
